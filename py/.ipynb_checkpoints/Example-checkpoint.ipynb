```python
import netCDF4
import numpy as np
import clima_anom as ca
import cartopy
import cartopy.crs as ccrs
import matplotlib.pyplot as plt
```

```python
hgt_dir = '../in/Hgt_500hPa_Jan80_Dec19.nc'
hgt_data = ca.read_netcdf(hgt_dir,2)
```

```python
lat = hgt_data['lat']
lon = hgt_data['lon']-180
tim = hgt_data['time']
hgt = hgt_data['hgt'][:,0,:,:]
```

```python
lon2d, lat2d = np.meshgrid(lon, lat)
```

```python
cmap = plt.cm.Spectral_r
proj = ccrs.PlateCarree()

plt.figure(figsize=(6,4),dpi=150)
ax = plt.axes(projection=proj)
ax.coastlines(linewidth=0.5)

a1 = ax.contourf(lon2d, lat2d, hgt[0,:,:], cmap = cmap)
gl = ax.gridlines(crs = proj,draw_labels=True,linewidth=0.6,color='gray',alpha=0.6,linestyle='--')
gl.top_labels = False
gl.bottom_labels = False
gl.right_labels = False
gl.xlabel_style = {'size': 8, 'color': 'black'}
gl.ylabel_style = {'size': 8, 'color': 'black'}
plt.colorbar(a1,orientation="horizontal")
```

```python
test = ca.remove_continent_ocean(hgt[:,:,:],lon,lat,remove='continent')
```

```python
cmap = plt.cm.Spectral_r
proj = ccrs.PlateCarree()

plt.figure(figsize=(6,4),dpi=150)
ax = plt.axes(projection=proj)
ax.coastlines(linewidth=0.5)

a1 = ax.contourf(lon2d, lat2d, test[0,:,:], cmap = cmap)
gl = ax.gridlines(crs = proj,draw_labels=True,linewidth=0.6,color='gray',alpha=0.6,linestyle='--')
gl.top_labels = False
gl.bottom_labels = False
gl.right_labels = False
gl.xlabel_style = {'size': 8, 'color': 'black'}
gl.ylabel_style = {'size': 8, 'color': 'black'}
plt.colorbar(a1,orientation="horizontal")
```
